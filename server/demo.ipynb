### Importing Packages

```python
import pandas as pd

file_path = "data/excel_file.xlsx"
```

### Read Excel Sheet

```python
df = pd.read_excel(
    file_path, 
    sheet_name="Apr 91",
    skiprows=2,
    usecols=[5, 6, 8],
)
```

### Change column names

```python
df.columns = ["Passed", "Job No.", "Code Head"]
```

### Perform Operation

```python
df = df.dropna(subset=["Passed", "Job No.", "Code Head"])
df
```

```python
df = df.groupby(["Code Head", "Job No."]).sum()
df
```

```python
def searchByJobNo(job: str) -> pd.DataFrame:
    return df.loc[df.index.get_level_values("Job No.") == job]


def searchByCodeHead(codeHead: str) -> pd.DataFrame:
    return df.loc[df.index.get_level_values("Code Head") == codeHead]
```
